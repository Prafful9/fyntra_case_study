```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = pd.DataFrame(dataset.iloc[:,3:7].values)
y = pd.DataFrame(dataset.iloc[:,7].values)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=85)

regressor = LinearRegression()
model = regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
plt.scatter(y_test, y_pred, color='red')
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('Test vs. Pred')
plt.grid(True)
plt.show()
```

```python
from sklearn import metrics

RMSE = round(np.sqrt(metrics.mean_squared_error(y_test,y_pred)),2)
print (RMSE)
```

```python
sns.distplot((y_test-y_pred), bins=50)
plt.show()
```

```python
col_coef = pd.DataFrame(regressor.coef_)
n=0
print ("Coefficient based on feature is as follows:\n")
for i in range(3,7):
    print (dataset.columns[i], ' = ', col_coef.iloc[0,n])
    n += 1
```
